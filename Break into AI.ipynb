## &rArr; Numpy Quick-start

First we will need to import the numpy library to our notebook

```python
import numpy as np
```

Below we will demo some matrix operations using python.

```python
# Create a 1D array with the elements 1, 2 & 3
a = np.array([1, 2, 3])

# Print the type of the variable "a"
print("Type of variable 'a': ", type(a))

# Print the shape (dimensions) of the array "a" in the form of (rows, columns)
print("Shape of array 'a': ", a.shape)
```

```python
# Create a 2D array with the elements 1, 2 & 3 in one row, 4, 5 & 6 in the other row
b = np.array([[1,2,3],[4,5,6]])

# Print the shape (dimensions) of the array "b" in the form of (rows, columns)
print("Shape of array 'b': ", b.shape)

# Select specific items from the array by indexing using its position
print("First item in the first row: ", b[0, 0])
print("Second item in the first row: ", b[0, 1])
print("First item in the second row: ", b[1, 0])
```

Let's try running some matrix calculations

```python
# Define our matrices (2D arrays)
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

# Elementwise sum; both produce the array
print("Sum of array 'x' and array 'y': ", np.add(x, y))
```

```python
print("Dot product of array 'x' and array 'y': ", np.dot(x, y))
```

```python
# Transpose the matrix 'x'
print ("The transpose of array 'x': ", (x.T))
```

## &rArr; SciPy Quick-start

We will do some Image manipulation here usin SciPy package. Images are mainly 3D-matrices/arrays of pixels information. An image consists of an array representing the height of an image, another array for representing the width of the image, and the last array for number of color channels in an image.

For example, if an image is 600x400px and is colored, it will have the following shape (400, 600, 3).

## &rArr; Matplotlib Quick-start

We will do some visualizations for our data here using pyplot function from matplotlib. Let's first import the function from the library.

```python
import matplotlib.pyplot as plt
```

We want to create a bar graph bow to describe the relationship between the streets of the houses in our dataset and the ratings of these houses.

```python
# Get index (position number) for the columns we need in the DataFrame
x_index = df_data_1.columns.get_loc('street')
y_index = df_data_1.columns.get_loc('review_scores_value')

print(x_index)
print(y_index)
```

```python
# Set the size of the plot
fig, ax = plt.subplots(figsize=(20, 10))

# Select the bar plot from pyplot, specify the columns that will be plotted on the x and y axes
plt.bar(np.array(df_data_1.iloc[:10, 34]), np.array(df_data_1.iloc[:10, 82]))

# Select the text values that will be shown on the x axis for different values (street names)
plt.xticks(list(df_data_1.iloc[:10, 34]), df_data_1.iloc[:10, 34])
# Rotate the x ticks to be nearly vertical
plt.xticks(rotation=80)

# Write text on the plot for defining x and y axes names
plt.xlabel('Street Name')
plt.ylabel('House Rating')

# Wrote text on the plot for defining the plot name (title)
plt.title('House Rating vs. Street/Location')

# Show the plot
plt.show()
```

Let's return now to our image example from the SciPy section. We want to view our image inside the notebook. That was not possible with SciPy, as it only shoes images in an external view.

We will use pyplot from matplotlib to view images, and we will continue manipulating images using SciPy and viewing the results with matplotlib.

```python
# Let's view our image (in the form of an array) that had the variable name "img"
fig, ax = plt.subplots(figsize=(20, 10))
plt.imshow(img)
```
